```python
#Importing Libraries
from google.colab import drive
import pandas as pd
import numpy as np
from numpy import mean, std
from pycountry_convert import country_name_to_country_alpha2
from pycountry_convert import country_alpha2_to_continent_code

drive.mount('/content/drive')
directory = '/content/drive/MyDrive/Data Projects/Happiness/Files'

pd.set_option("display.max_rows", None, "display.max_columns", None)
```

```python
#Importing Data
df_2009 = pd.read_csv(f'{directory}/World Happiness All.csv')
df_2021 = pd.read_csv(f'{directory}/World Happiness 2021.csv')
```

```python
#Dropping Unused Data
df_2009 = df_2009[df_2009.year == 2009]
df_2009 = df_2009.drop(df_2009.columns[[1, 9,  10]], axis=1)
df_2009 = df_2009.drop(907)
df_2021 = df_2021.drop(df_2021.columns[[1, 3, 4, 5, 12, 13, 14, 15, 16, 17, 18, 19]], axis=1)
df_2021 = df_2021.drop(32)

#Standardizing Feature Names
df_2009 = df_2009.rename(columns={'Log GDP per capita': 'Logged GDP per capita', 'Life Ladder': 'Happiness score', 'Healthy life expectancy at birth': 'Healthy life expectancy'})
df_2021 = df_2021.rename(columns={'Ladder score': 'Happiness score'})

#Standardizing Country Names
df_2021.loc[23,'Country name']='Taiwan'
df_2021.loc[73,'Country name']='Cyprus'
df_2021.loc[76,'Country name'] = 'Hong Kong'
df_2009.loc[701,'Country name'] = 'Hong Kong'
df_2021.loc[82,'Country name'] = 'Democratic Republic of the Congo'
df_2009.loc[388,'Country name'] = 'Democratic Republic of the Congo'
df_2021.loc[93,'Country name'] = 'Macedonia'
df_2009.loc[1288,'Country name'] = 'Macedonia'
df_2021.loc[124,'Country name'] = 'Palestine'
df_2009.loc[1328,'Country name'] = 'Palestine'
df_2009.loc[1562,'Country name'] = 'Somalia'	
```

## Data Cleaning

```python
#Checking For Missing Values
missing_2009 = df_2009.isnull().sum()
print('Missing Values (2009)')
print(missing_2009)
missing_2021 = df_2021.isnull().sum()
print('Missing Values (2021)')
print(missing_2021)
```

```python
#Mean imputing missing values for 2009 data
df_2009['Logged GDP per capita']=df_2009['Logged GDP per capita'].fillna(df_2009['Logged GDP per capita'].mean())
df_2009['Social support']=df_2009['Social support'].fillna(df_2009['Social support'].mean())
df_2009['Healthy life expectancy']=df_2009['Healthy life expectancy'].fillna(df_2009['Healthy life expectancy'].mean())
df_2009['Freedom to make life choices']=df_2009['Freedom to make life choices'].fillna(df_2009['Freedom to make life choices'].mean())
df_2009['Generosity']=df_2009['Generosity'].fillna(df_2009['Generosity'].mean())
df_2009['Perceptions of corruption']=df_2009['Perceptions of corruption'].fillna(df_2009['Perceptions of corruption'].mean())

#Rechecking For Missing Values
missing_2009 = df_2009.isnull().sum()
print('Missing Values (2009):')
print(missing_2009)
```

```python
#Adding Features for Country Names, Country Codes, and Continent Codes (2009)
for index, rows in df_2009.iterrows():
    cn_name = df_2009.loc[index,'Country name']
    try:
        df_2009.loc[index,'Country code'] = country_name_to_country_alpha2(cn_name, cn_name_format="default")
        cn_code = df_2009.loc[index,'Country code']
        df_2009.loc[index,'Continent code'] = country_alpha2_to_continent_code(cn_code)
    except:
        df_2009.loc[index,'Country code'] = 'unknown'
        df_2009.loc[index,'Continent code'] = 'unknown'
    if df_2009.loc[index, 'Continent code'] == 'NA':
        df_2009.loc[index,'Continent'] = 'North America'
    elif df_2009.loc[index, 'Continent code'] == 'SA':
        df_2009.loc[index,'Continent'] = 'South America'
    elif df_2009.loc[index, 'Continent code'] == 'AS':
        df_2009.loc[index,'Continent'] = 'Asia'
    elif df_2009.loc[index, 'Continent code'] ==  'OC':
        df_2009.loc[index,'Continent'] = 'Australia/Oceania'
    elif df_2009.loc[index,'Continent code'] == 'EU':
        df_2009.loc[index,'Continent'] = 'Europe'
    elif df_2009.loc[index,'Continent code'] == 'AF':
        df_2009.loc[index,'Continent'] = 'Africa'
    elif df_2009.loc[index,'Continent code'] == 'AN':
        df_2009.loc[index,'Continent'] = 'Antarctica'

#Adding Features for Country Names, Country Codes, and Continent Codes (2021)
for index, rows in df_2021.iterrows():
    cn_name = df_2021.loc[index,'Country name']
    try:
        df_2021.loc[index,'Country code'] = country_name_to_country_alpha2(cn_name, cn_name_format="default")
        cn_code = df_2021.loc[index,'Country code']
        df_2021.loc[index,'Continent code'] = country_alpha2_to_continent_code(cn_code)
    except:
        df_2021.loc[index,'Country code'] = 'unknown'
        df_2021.loc[index,'Continent code'] = 'unknown'
    if df_2021.loc[index, 'Continent code'] == 'NA':
        df_2021.loc[index,'Continent'] = 'North America'
    elif df_2021.loc[index, 'Continent code'] == 'SA':
        df_2021.loc[index,'Continent'] = 'South America'
    elif df_2021.loc[index, 'Continent code'] == 'AS':
        df_2021.loc[index,'Continent'] = 'Asia'
    elif df_2021.loc[index, 'Continent code'] ==  'OC':
        df_2021.loc[index,'Continent'] = 'Australia/Oceania'
    elif df_2021.loc[index,'Continent code'] == 'EU':
        df_2021.loc[index,'Continent'] = 'Europe'
    elif df_2021.loc[index,'Continent code'] == 'AF':
        df_2021.loc[index,'Continent'] = 'Africa'
    elif df_2021.loc[index,'Continent code'] == 'AN':
        df_2021.loc[index,'Continent'] = 'Antarctica'

#Adding year feature
df_2009['Year'] = '2009'
df_2021['Year'] = '2021'
```

```python
df_2009.to_csv(f'{directory}/2009_Happiness.csv') 
df_2021.to_csv(f'{directory}/2021_Happiness.csv') 
```
